```python
#16-1
import csv
from matplotlib import pyplot as plt
from datetime import datetime
filename = 'death_valley_2014.csv'
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)
    #print(header_row)
    highs,lows,dates = [],[],[]
    for row in reader:
        try:
            #print(row)
            current_date = datetime.strptime(row[0],"%Y-%m-%d")
            high = int(row[1])
            low = int(row[3])
        except:
            print(current_date,' missing data')
        else:
            highs.append(high)
            lows.append(low)
            dates.append(current_date)
    fig = plt.figure(dpi=128,figsize=(10,6))
    plt.plot(dates,highs,c='red')
    plt.plot(dates,lows,c='blue')
    plt.fill_between(dates,highs,lows,facecolor='blue',alpha=0.1)
    plt.title("Daily high and low temperature -2014",fontsize=15)
    plt.show()
```

```python
import pygal
filename = 'gdp_json.json'
with open(filename) as f:
    gdp_data = json.load(f)
years,gdpvalues=[],[]
for gdp_dict in gdp_data:
    cname = gdp_dict['Country Name']
    if cname=='China':
        gdpvalues.append(int(float(gdp_dict['Value']))/100000000)
        years.append(gdp_dict['Year'])
line_chart = pygal.Line(x_label_rotation=20, show_minor_x_labels=False)
line_chart.title = '中国历年GDP'
N = 3
line_chart.x_labels = years
line_chart.x_labels_major = years[::N] #切片操作基本表达式：object[start_index:end_index:step]
line_chart.add('GDP（亿元）',gdpvalues)
line_chart.render_to_file('中国GDP折线图(亿元).svg')
```

```python
import pygal
import math
line_chart = pygal.Line(x_label_rotation=20, show_minor_x_labels=False)
line_chart.title = '中国历年GDP对数变换'
N = 3
line_chart.x_labels = years
line_chart.x_labels_major = years[::N] #切片操作基本表达式：object[start_index:end_index:step]
gdplog_val = [math.log10(a) for a in gdpvalues]
line_chart.add('logGDP（亿元）',gdplog_val)
line_chart.render_to_file('中国GDP对数折线图(亿元).svg')
```
