```python
#7-5
prompt='please input your age to calculate the fee for ticket \n'
prompt+="enter 'quit' to end："
while True:
    age = input(prompt)
    if age=='quit':
        print('end')
        break;
    intage = int(age)
    if intage<3:
        print('free')
    elif intage>=3 and intage<12:
        print('10 美元')
    else:
        print('15 美元')
```

```python
#7-6 
prompt='please input your age to calculate the fee for ticket \n'
prompt+="enter 'quit' to end："
active=True
while active:
    age = input(prompt)
    if age=='quit':
        active = False
        break
    else:
        if int(age)<3:
            print('free')
        elif int(age)>=3 and int(age)<12:
            print('10 美元')
        else:
            print('15 美元')
```

```python
#7-8
sandwich_orders=['tuna','pork','egg','pastrami']
finished_sandwich=[]
while sandwich_orders:
    san = sandwich_orders.pop()
    print('I made your ' + san + ' sandwich')
    finished_sandwich.append(san)
print('Have finished all sandwich:')
for sandwich in finished_sandwich:
    print(sandwich)
```

```python
#7-9 修改列表中的内容，不能用for循环，修改了只有无法for each
sandwich_orders=['tuna','pork','egg','pastrami','pastrami','pastrami','pastrami']
print('pastrami is selled out')
while 'pastrami' in sandwich_orders:
    sandwich_orders.remove('pastrami')
for sandwich in sandwich_orders:
    print(sandwich)
```

```python
#7-10
travel={}
while True:
    name=input('what is your name?')
    place=input('If you could visit one place in the world, where would you go?')
    travel[name]=place
    response=input('Would you like to let another person respond?(yes/no)')
    if response=='no':
        break;
for name,place in travel.items():
    print(name.title() + ' would like to go to ' + place.title())
```
